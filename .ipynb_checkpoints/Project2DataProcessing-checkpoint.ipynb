```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Counting Ones

```python
ones = pd.read_csv('countingOnes.csv', names=['Algorithm','N-bits', 'Ones'])
```

```python
ones.head()
```

```python
ones.groupby(['Algorithm', 'N-bits']).mean().unstack(level=0).plot(title="Counting Ones Optimal Solution")
plt.ylabel("Number of Ones")
```

# Traveling Sales Man

```python
tsm = pd.read_csv('tsm.csv', names=['Algorithm','Nodes','1/Distance'])
tsm.head()
```

```python
tsm.groupby(['Algorithm','Nodes']).mean().unstack(level=0).plot(title='Traveling Salesman Optimization')
plt.ylabel('1/Distance')
```

# SA

```python
sa_cooling = pd.read_csv('sa_cooling.csv')
sa_cooling.head()
```

```python
avg = sa_cooling[['Trial1','Trial2','Trial 3','Trial 4 ','Trial 5']].mean(axis=1)
```

```python
test = pd.concat([sa_cooling[['Cooling','Iter']],avg], axis=1)
```

```python
ax = (test.loc[test.Cooling == 0.15]).plot(x='Iter',y=0,label='0.15')
test.loc[test.Cooling == 0.95].plot(ax =ax, x='Iter',y=0, label = '0.95')
test.loc[test.Cooling == 0.55].plot(ax=ax, x='Iter',y=0, label = '0.55')
plt.ylim([0,100])
plt.title('Simulated Annealing at Various Cooling Rates')
plt.ylabel('Accuracy %')
```

```python
sa_temp = pd.read_csv('sa_temp.csv')
sa_temp.head()
```

```python
avg_temp = sa_temp[['Trial 1','Trial 2','Trial 3','Trial 4','Trial 5']].mean(axis=1)
```

```python
temp = pd.concat([sa_temp[['Temp','Iter']],avg_temp], axis=1)
```

```python
ax = (temp.loc[temp.Temp == 100000000000]).plot(x='Iter',y=0,label='1E11')
temp.loc[temp.Temp == 1000].plot(ax =ax, x='Iter',y=0, label = '1000')
temp.loc[temp.Temp == 1].plot(ax=ax, x='Iter',y=0, label = '1')
plt.ylim([0,100])
plt.title('Simulated Annealing at Various Temperatures')
plt.ylabel('Accuracy %')
```

```python
r10 = pd.read_csv('mutationRate10.csv')
r50 = pd.read_csv('mutationRate50.csv')
r90 = pd.read_csv('mutationRate90.csv')
```

```python
mutationRate = pd.DataFrame()
```

```python
mutationRate['10%'] = r10.mean(axis=1)
mutationRate['50%'] = r50.mean(axis=1)
mutationRate['90%'] = r90.mean(axis=1)
```

```python
mutationRate.plot(title="Comparing Flappy Bird Mutation Rates")
plt.ylabel('End Score')
plt.xlabel('Generation')
```

```python
point = pd.read_csv('singlePointCrossover.csv')
one = pd.read_csv('oneCrossOver.csv')
two = pd.read_csv('twoCrossOver.csv')
```

```python
crossOver = pd.DataFrame()
```

```python
crossOver['Single Weight One-Point Crossover'] = point.mean(axis=1)
crossOver['Multi-weight One-Point Crossover'] = one.mean(axis=1)
crossOver['Two-Point Crossover'] = two.mean(axis=1)
```

```python
crossOver.plot(title="Comparing Crossover Methods in Flappy Bird")
plt.ylabel('End Score')
plt.xlabel('Generation')
```
